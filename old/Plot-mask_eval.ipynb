```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\ComputerScience\Mestrado\results\UNIFESP360\dice_scores.csv')
df2 = pd.read_csv(r'D:\ComputerScience\Mestrado\results\Occlusion\dice_scores.csv')
```

```python
colors = ['pink', 'lightblue', 'lightgreen']
fig1, ax1 = plt.subplots(1,1, figsize=(7,5), sharey=True)
fig2, ax2 = plt.subplots(1,1, figsize=(7,5), sharey=True)

axs = [ax1, ax2]

data = df.to_numpy()[:,1:]
bplot = axs[0].boxplot(data, vert=True, patch_artist=True, labels=df.columns.values[1:])
axs[0].set_title('UNIFESP Dataset')

data = df2.to_numpy()[:,1:]
bplot2 = axs[1].boxplot(data, vert=True, patch_artist=True, labels=df.columns.values[1:])
axs[1].set_title('Occlusion Dataset')

for ax in axs:
    for bplot in [bplot, bplot2]:
        for patch, color in zip(bplot['boxes'], colors):
            patch.set_facecolor(color)
    ax.yaxis.grid(True)

ax0_legend = [
    f"={df['landmark_seg'].median():.2f}; ={df['landmark_seg'].mean():.2f}; ={df['landmark_seg'].std():.2f}", 
    f"={df['deeplabv3p'].median():.2f}; ={df['deeplabv3p'].mean():.2f}; ={df['deeplabv3p'].std():.2f}",
    f"={df['sam'].median():.2f}; ={df['sam'].mean():.2f}; ={df['sam'].std():.2f}"]
ax1_legend = [
    f"={df2['landmark_seg'].median():.2f}; ={df2['landmark_seg'].mean():.2f}; ={df2['landmark_seg'].std():.2f}", 
    f"={df2['deeplabv3p'].median():.2f}; ={df2['deeplabv3p'].mean():.2f}; ={df2['deeplabv3p'].std():.2f}",
    f"={df2['sam'].median():.2f}; ={df2['sam'].mean():.2f}; ={df2['sam'].std():.2f}"]
axs[0].set_ylabel('Dice Score')
axs[0].legend(bplot["boxes"], ax0_legend, loc='lower right')
#axs[0].set_xlim(0,5)
axs[1].set_ylabel('Dice Score')
axs[1].legend(bplot2["boxes"], ax1_legend, loc='lower right')
#axs[1].set_xlim(0,5)
```
